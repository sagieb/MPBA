```python
contexts = ['Aim18', 'Gpx1', 'Gsy2', 'ND', 'Stp4']
genome_exp = '30'
plasmid_exp_num = '13'
context_exp = '28'
exps = [genome_exp,plasmid_exp_num,context_exp]
tp='180'
tf= 'Msn2'
context_files = [file for file in os.listdir(params.RES_PATH) if file.split('.')[0].split('_')[-1]==context_exp]
genome_files = [file for file in os.listdir(params.RES_PATH) if file.split('.')[0].split('_')[-1]==genome_exp]
genome_libs = np.unique([file.split('_')[1] for file in os.listdir(params.RES_PATH) if file.split('.')[0].split('_')[-1]==genome_exp and file.split('_')[0]==tf])

norm_seq_by_lib = {}
context_files = [file.split('_')[1] for file in context_files]
for lib_num in context_files:
    lib_info = dp.get_lib_info(lib_num)    
    norm_seqs_lib = pd.DataFrame()    
    for context in contexts:
        norm_seqs_lib[context] = dp.mean_over_bio_reps(lto.norm_res_data(tf+'Context'+context, lib_num, context_exp, norm_to=None, occ=None))[tp]
    if lib_num in genome_libs:
        norm_seqs_lib['Genome'] = dp.mean_over_bio_reps(lto.norm_res_data(tf, lib_num, genome_exp, norm_to=None, occ=None))[tp]
    norm_seqs_lib['Plasmid'] = dp.mean_over_bio_reps(lto.norm_res_data(tf, lib_num, plasmid_exp_num, norm_to=None, occ=None))[tp]
    norm_seq_by_lib[lib_num] = norm_seqs_lib
```

## (A) occupcncy at wild type sequence - Msn2 in contexts

```python
tf='Msn2'
wt_var_seq_libs = pd.DataFrame()
for lib_num in norm_seq_by_lib.keys():
    if lib_num =='86':
        continue
    lib_info = dp.get_lib_info(lib_num)
    wt_var_seq = ''.join(lib_info['wt_at_var_loc'])
    lib_wt_contexts = pd.DataFrame()
    for context in contexts:
        tf_context = tf+'Context'+context
        norm_df_occ = lto.norm_without_biorep_mean(tf_context, lib_num, context_exp, None, occ=True) *100 # normlize data
        context_wt_vals = norm_df_occ.loc[wt_var_seq]
        mean_context_wt_val = context_wt_vals.mean()
        sem_context_wt_val = context_wt_vals.sem()
        context_df = pd.DataFrame({'Library':[lib_info['gene']], 'lib_num':[lib_num], 'Context':[context],
                                   'wt_val':[mean_context_wt_val],'wt_sem':[sem_context_wt_val] })
        lib_wt_contexts = pd.concat([lib_wt_contexts,context_df])
    wt_var_seq_libs = pd.concat([wt_var_seq_libs, lib_wt_contexts])

result = wt_var_seq_libs.groupby(["Library"])['wt_val'].aggregate(np.median).reset_index().sort_values('wt_val').iloc[::-1]
f,ax = plt.subplots(1,1,figsize=(8,3))
sns.barplot(data=wt_var_seq_libs, x='Library', y='wt_val', hue='Context', order=result['Library'], ax=ax)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1), ncol=1)

x_coords = [p.get_x() + 0.5*p.get_width() for p in ax.patches][0:-5]
y_coords = [p.get_height() for p in ax.patches][0:-5]
ax.errorbar(x=x_coords, y=y_coords, yerr=wt_var_seq_libs["wt_sem"], fmt="none", c= "k")
```

## (B) Correlation in contexts

```python
fig, axes = plt.subplots(2,4,figsize=(16,7),constrained_layout=True)
cmap= sns.color_palette("crest_r", as_cmap=True)
for lib_i, lib_num in enumerate(norm_seq_by_lib.keys()):
    curr_ax = axes.ravel()[lib_i]
    lib_info = dp.get_lib_info(lib_num)
    norm_lib_data = norm_seq_by_lib[lib_num]
    norm_lib_data_corr = norm_lib_data.corr()
    sns.heatmap(norm_lib_data_corr, ax=curr_ax, cmap=cmap)
    curr_ax.set_title(lib_info['gene'])
```
