```python
exp = 30
truncs = ['T1','T2','T3','T4','T5','T6','TU','TD']
tfs = ['Msn2','Msn2DBDOE','Msn2NONDBDOE']
libs = [32,56,72,87,88]
norm_to = 10**6

for tf in tfs:
    for trunc in truncs:
        tfc = tf + 'ContextStp4' + trunc
        for lib in libs:
            filtered_reads = dp.filter_results(tfc,lib,exp)
            read_normed = filtered_reads.div(filtered_reads.sum())*norm_to
            fig,ax = plt.subplots(1,1)
            
            sns.heatmap(read_normed.corr(),ax=ax)
            ax.set_title(tfc + ' ' + str(lib))
            fig.savefig('corrs/'+tfc+'_'+lib+'.png')
            fig.close()
# nr = dp.norm_reads(dp.filter_results(tf,lib,exp))
# samp_info = dp.get_samp_info(nr)
# nar = dp.mean_bio_rep_tps(nr,samp_info).apply(np.log2)
# samp_info = dp.get_samp_info(nar)
# nnn = dp.norm_to_tp_0(nar,samp_info)
# sns.heatmap(nr.corr())
# frfr.sum()
```
